```python
import pandas as pd
import ipywidgets as widgets
```

```python
url = "https://docs.google.com/spreadsheets/d/1jLNfP3iuteUJrH0zS9qWONskyKh9pFcl1hKSlgEc-I8/gviz/tq?tqx=out:csv&sheet=All_data"
```

```python
all_data = pd.read_csv(url)
```

```python
# Create a string var that will be the legend in the source comparison plots

all_data['source_legend'] = all_data['short_reference'] + " - " + all_data['welfare_concept']

all_data.loc[all_data['series_code'].str[0] == "F", 'source_legend'] = "Chartbook series"
```

```python
dropdown_country = widgets.Dropdown(options = all_data.country.unique().tolist())

# An HBox lays out its children horizontally
ui = widgets.HBox([dropdown_country])

def print_data(dropdown_country):
    filtered_df = all_data[all_data.country == dropdown_country]
    display(filtered_df.head())

out = widgets.interactive_output(print_data, {'dropdown_country': dropdown_country})

display(ui, out)
```
